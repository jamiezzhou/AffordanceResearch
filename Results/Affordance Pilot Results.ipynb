```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
carolyn = pd.read_csv("Carolyn.csv")
eyeHeight = carolyn['eyeHeight'][0]

carolyn['affordance'] = carolyn['affordance'].apply(lambda x: x/eyeHeight)

#import carolyn avatar = 0, type = 0
carolyn1 = carolyn[carolyn['avatar'] == 0]
carolyn1 = carolyn1[carolyn1['type'] == 0]

#import carolyn avatar = 0, type = 1
carolyn2 = carolyn[carolyn['avatar'] == 0]
carolyn2 = carolyn2[carolyn2['type'] == 1]

#import carolyn avatar = 1, type = 0
carolyn3 = carolyn[carolyn['avatar'] == 1]
carolyn3 = carolyn3[carolyn3['type'] == 0]

#import carolyn avatar = 1, type = 1
carolyn4 = carolyn[carolyn['avatar'] == 1]
carolyn4 = carolyn4[carolyn4['type'] == 1]

x = ['Pretest','Block1','Block2','Posttest']
avg1 = carolyn1['affordance'].groupby(carolyn1['part']).mean()
avg2 = carolyn2['affordance'].groupby(carolyn2['part']).mean()
avg3 = carolyn3['affordance'].groupby(carolyn3['part']).mean()
avg4 = carolyn4['affordance'].groupby(carolyn4['part']).mean()

plt.figure(figsize = (6,4))
plt.errorbar(x, avg1, label = 'Avatar = 0, Type = 0', marker='o', linestyle='dashed')
plt.errorbar(x, avg2, label = 'Avatar = 0, Type = 1', marker='o', linestyle='dashed')
plt.errorbar(x, avg3, label = 'Avatar = 1, Type = 0', marker='o', linestyle='dashed')
plt.errorbar(x, avg4, label = 'Avatar = 1, Type = 1', marker='o', linestyle='dashed')

avg = carolyn['affordance'].groupby(carolyn['part']).mean()
print(avg)
plt.errorbar(x, avg, label = 'Average', marker='o', linestyle='dashed')

plt.legend()
plt.title('Step Over Affordance: Participant 1')
plt.xlabel('Experiment Phase')
plt.ylabel('Affordance Judgement Eye Height Ratio')
```
